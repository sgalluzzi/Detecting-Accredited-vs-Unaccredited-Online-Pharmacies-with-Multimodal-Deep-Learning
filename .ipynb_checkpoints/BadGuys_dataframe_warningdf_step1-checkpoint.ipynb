```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.fda.gov/drugs/drug-supply-chain-integrity/internet-pharmacy-warning-letters'
data = requests.get(url)
soup = BeautifulSoup(data.text, 'html.parser')
```

```python
print(data)
```

```python
bg = soup.find_all(["tbody","th","title"])
```

```python
big = soup.find_all('td headers="header1"')
```

```python
print(bg)
```

```python
bad_guys = []
for b in bg[0:]:
    result = b.text.strip()
    bad_guys.append(result)
```

```python
print(bad_guys)
```

```python
df = pd.DataFrame (bad_guys, columns = ['column_name'])
```

```python
df
```

```python
df = df.drop([0,1,2,3])
```
