```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.fda.gov/drugs/drug-supply-chain-integrity/internet-pharmacy-warning-letters"
headers = {'user-agent': 'ds6050 (fdf7gn@virginia.edu)'}
r = requests.get(url, headers=headers)
r
```

```python
fda = BeautifulSoup(r.text, 'html')
```

```python
fda_web = "https://www.fda.gov"
table2 = fda.find("div", "table-responsive")
table_atags = table2.find_all("a")
warn_url = [fda_web + i['href'] if i['href'].startswith('/') else i['href'] for i in table_atags ]
warn_url = warn_url[2:]
warn_url = warn_url[:-3]
warn_url
```

```python
def recipient(url):
  headers = {'user-agent': 'ds6050 (fdf7gn@virginia.edu)'}
  r = requests.get(url, headers=headers)
  warn_ltr = BeautifulSoup(r.text, 'html')
  to = warn_ltr.find("div", "col-xs-12 col-md-6")
  dd = to.find_all("dd")
  web_url = dd[1].text.split('>')
  return web_url

```

```python
web_list = []
for url in warn_url:
  #print(url)
  webs = recipient(url)
  web_list.append(webs)
web_list
```

```python
def tableweb(url):
  headers = {'user-agent': 'ds6050 (fdf7gn@virginia.edu)'}
  r = requests.get(url, headers=headers)
  warn_ltr = BeautifulSoup(r.text, 'html')
  tab = warn_ltr.find_all("td")
  url_list = []
  if tab:
    for i in tab:
      url_list.append(i.text.strip())
  else:
    pass
    
  web_url = url_list[1:]
  #print(web_url)
  return web_url
```

```python
for url in warn_url:
  #print(url)
  webs = tableweb(url)
  web_list.append(webs)
```

```python
h = 'https://'
d2 = [h + i for i in data ]


titles=[]
def scraper(link):
    try:
        from urllib.request import urlopen
        from bs4 import BeautifulSoup
        
        htmld = urlopen(link)
        headers = {'user-agent': 'ds6050 (fdf7gn@virginia.edu)'}
        r = requests.get(link, headers=headers)
        soup = BeautifulSoup(htmld,'html.parser')
        body = soup.find_all('title')
        for tag in body:
            titles.append(tag.text.strip())
            
    except:
        titles.append("This one is a 403 error ---- " + link)
        pass

    return (titles)

mainDataFrame = pd.DataFrame()

for i in d2:
    dictObj =scraper(i)
    mainDataFrame = mainDataFrame.append(dictObj)
```
